```python
import pandas as pd
import numpy as np
import math
import regex as re
```

### DATA MUNGING

```python
##Reading the data to get tourism indicators 
tourism_indicator= pd.read_csv('../data/host_tourism_indicators.csv')
```

```python
##Checking how the data looks
tourism_indicator.head()
```

```python
##columns
tourism_indicator.columns
```

```python
#Info
tourism_indicator.info()
```

```python
##Replacing irrelevant strings with null values
tourism_indicator=tourism_indicator.replace({"\\.\\.":np.nan},regex=True)
```

```python
tourism_indicator.info()
```

```python
##Removing variables that have more than 60% data missing
removed_columns_null=[]
for i in tourism_indicator.columns:
    if tourism_indicator[i].isnull().sum()>(0.6*len(tourism_indicator)):
        removed_columns_null.append(i)
tourism_indicator=tourism_indicator.drop(removed_columns_null,axis=1)

print(tourism_indicator.info())
```

```python
tourism_indicator.columns
```

```python
tourism_indicator.head()
```

```python
##keeping relevant columns in the dataset
tourism_indicator=tourism_indicator.drop(['Series Name', 'Series Code'],axis=1)
```

```python
tourism_indicator.head()
```

```python
##Removing irrelavant rows from the dataset
tourism_indicator=tourism_indicator[(tourism_indicator['Country Name']!= 'Data from database: World Development Indicators')]
tourism_indicator=tourism_indicator[(tourism_indicator['Country Name']!= 'Last Updated: 04/08/2022') ]
tourism_indicator=tourism_indicator.dropna(subset=['Country Name'])
```

```python
tourism_indicator.head()
```

```python
##Tidying data to ready for the visualisation
tourism_indicator=pd.melt(tourism_indicator, id_vars=['Country Name', 'Country Code'], value_vars=['1995 [YR1995]', '1996 [YR1996]',
       '1997 [YR1997]', '1998 [YR1998]', '1999 [YR1999]', '2000 [YR2000]',
       '2001 [YR2001]', '2002 [YR2002]', '2003 [YR2003]', '2004 [YR2004]',
       '2005 [YR2005]', '2006 [YR2006]', '2007 [YR2007]', '2008 [YR2008]',
       '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]',
       '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]',
       '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]'], var_name='year', value_name='tourism_arrivals', col_level=None)
```

```python
##Cleaning values for the year column
tourism_indicator['year'] = tourism_indicator['year'].apply(lambda x: re.sub(r"\[......\]", "", x))
```

```python
tourism_indicator.head()
```

```python
##Converting all the variables to suitable type
tourism_indicator=tourism_indicator.astype({'year': 'int32','tourism_arrivals': 'float64'})
tourism_indicator.rename({"Country Name":"host_country",
                "Country Code":"country_code"},axis=1,inplace=True)
tourism_indicator.dtypes
```

```python
##Imputing null values by the mean of the country's tourist arrivals population
tourism_indicator["tourism_arrivals"] = tourism_indicator.groupby("host_country")['tourism_arrivals'].transform(lambda x: x.fillna(x.mean()))
```

```python
##Storing our cleaned data into a file
tourism_indicator.to_csv('../data/tourism_clean.csv',index=False)
```

```python
##Preparing data to merge with our cleaned tourist data
event_data=pd.read_csv('../data/athlete_events.csv')
```

```python
event_data.head()
```

```python
##Creating host countries for each variable
def host_country(col):
    if col == "Rio de Janeiro":
        return "Brazil"
    elif col == "London"or col=='Sarajevo':
        return "United Kingdom"
    elif col == "Beijing":
        return  "China"
    elif col == "Athina":
        return  "Greece"
    elif col == "Sydney" or col == "Melbourne":
        return  "Australia"
    elif col == "Atlanta" or col == "Los Angeles" or col == "St. Louis" or col=='Squaw Valley'or col=='Salt Lake City'or col=='Lake Placid' or col=='Albertville':
        return  "United States"
    elif col == "Barcelona":
        return  "Spain"
    elif col == "Seoul":
        return  "Korea, Rep."
    elif col == "Moskva"or col=='Sochi':
        return  "Russian Federation"
    elif col == "Montreal"or col=="Vancouver"or col=='Calgary':
        return  "Canada"
    elif col == "Munich" or col == "Berlin"or col=='Garmisch-Partenkirchen':
        return  "Germany"
    elif col == "Mexico City":
        return  "Mexico"
    elif col == "Tokyo" or col=='Sapporo'or col=='Nagano':
        return  "Japan"
    elif col == "Roma" or col=='Torino'or col=="Cortina d'Ampezzo":
        return  "Italy"
    elif col == "Paris"or col=='Grenoble'or col=='Chamonix':
        return  "France"
    elif col == "Helsinki":
        return  "Finland"
    elif col == "Amsterdam":
        return  "Netherlands"
    elif col == "Antwerpen":
        return  "Belgium"
    elif col == "Stockholm":
        return  "Sweden"
    elif col=='Oslo' or col=='Lillehammer':
        return 'Norway'    
    elif col=='Sankt Moritz':
        return  "Switzerland"
    elif col=='Innsbruck':
        return  "Austria"
    else:
        return "Other"
```

```python
event_data['host_country'] = event_data['City'].apply(host_country)
```

```python
##As the tourism indicators are available from 1995, keeping all the data from 1995 and 2020 in our athlete events
##the focus is on summer olympics alone, hence removing winter olympics from the dataset
new_data=event_data.loc[(event_data.Year>=1995) &(event_data.Year<=2020) &(event_data.Season=='Summer') ]
```

```python
## giving appropriate names to columns
new_data.rename({"Year":"year"},axis=1,inplace=True)
```

```python
new_data.head()
```

```python
##Keeping relevant columns for analysis
new_data=new_data[['host_country','year']]
new_data['host']='Yes'
```

```python
##Merging our datasets (tourism indicator and athlete events) for analysis
joining_data = pd.merge(tourism_indicator, new_data,  how='outer', left_on=['host_country','year'], right_on = ['host_country','year'])
joining_data=joining_data.drop_duplicates()
```
